# Stock Price Prediction Utilizing LSTM Techniques
Python code for predicting stock prices using Long Short-Term Memory (LSTM) neural networks. The code imports financial data, preprocesses it, builds an LSTM model, and makes predictions on the stock's closing prices.


It's important to acknowledge that applying LSTM directly to raw stock price data can be challenging and impractical. Attempting to scale such data using a min-max scaler may not yield meaningful results because raw stock prices often lack stationarity and are not suitable for straightforward extrapolation. In practice, this approach may seem deceptively accurate, but it essentially creates a delayed curve rather than making meaningful predictions.

When working with LSTM for financial data prediction, a more reasonable approach is to focus on forecasting "Log Returns." This method is better aligned with financial modeling principles and provides a more stable foundation for making predictions. It's essential to understand that this project is better suited as a programming example for basic machine learning techniques rather than a precise stock price prediction tool. The emphasis here is on demonstrating the use of LSTM and data preprocessing techniques for educational purposes rather than delivering actionable stock price forecasts.

## Data Preprocessing

### Splitting Data into Training and Validation Sets

```python
Data['Date'] = pd.to_datetime(Data['Date'])
Data.set_index('Date',inplace=True)
Close = Data.filter(['Close'])
CloseValue = Close.values
TrainingDataLength = math.ceil(len(CloseValue)*.8)
TrainingDataLength
```

###  Scaling Data

```python
scaler = MinMaxScaler(feature_range=(0,1))
PriceData = scaler.fit_transform(CloseValue)
PriceData
## Customized the function:
# def Rank(data):
#     feature_range = data.max() - data.min()
#     scaled_data = (data - data.min()) / feature_range
#     return scaled_data
# PriceData = Rank(CloseValue)
# Rank(CloseValue)
```

### Creating Sequences

```python
X_train, Y_train = [],[]
Backcandles = 60
TrainData = PriceData[0:TrainingDataLength]
for i in range(Backcandles,len(TrainData)):
    X_train.append(TrainData[i-Backcandles:i, 0])
    Y_train.append(TrainData[i,0])
    if i<= Backcandles:
        print("X_train:",X_train,"\nY_train:",Y_train)
X_train,Y_train = np.array(X_train), np.array(Y_train)
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_train.shape
```

## LSTM Model Building, Compiling, and Training

```python
Model = Sequential([
    LSTM(50, return_sequences = True, input_shape = (X_train.shape[1], 1)),
    (Dropout(0.2)),
    LSTM((50)),
    (Dropout(0.2)),
    (Dense(32)),
    (Dense(1))
])

Model.compile(optimizer="adam", loss="mean_squared_error")
Model.fit(X_train, Y_train, batch_size=32, epochs=10)
Model.summary()
```

```python
test_data= PriceData[TrainingDataLength-Backcandles:, :]
x_test, y_test = [], CloseValue[TrainingDataLength:,:]
for i in range(Backcandles,len(test_data)):
    x_test.append(test_data[i-Backcandles:i,0])
x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1],1))
x_test.shape
```

## Results of the Prediction

### Root-Mean-Square Error
&emsp;&emsp;A higher RMSE value generally indicates poorer predictive performance. Hence, our training objective is to "minimize RMSE".

```python
Pred = Model.predict(x_test)
Pred = scaler.inverse_transform(Pred)
RMSE = np.sqrt(np.mean(Pred - y_test)**2)
RMSE
```

### Prediction Results

```python
TrainingSet,ValidationSet = Close[:TrainingDataLength],Close[TrainingDataLength:]
ValidationSet["Predictions"] = Pred
ValidationSet
```

```python
plt.figure(figsize=(12,8))
plt.title("Stock Price Prediction During the Validation Period", fontsize = 20)
plt.ylabel("Close Price")
plt.plot(ValidationSet["Close"],linewidth=1,color = "Darkviolet")
plt.plot(ValidationSet["Predictions"],linewidth=1,color = "Red")
plt.legend(["Stock Price","Predictions"])
```
